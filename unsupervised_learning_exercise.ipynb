```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Loading the dataset
df = pd.read_csv('customer_data_records.csv')
```

```python
# Selecting features for clustering
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
# Perform K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
```

```python
df['Cluster']=kmeans.labels_ # Assign cluster labels to the DataFrame
df
```

```python
# Visualize the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='viridis', s=100)
plt.title('Customer Segments based on Annual Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
#Another cluster visualization
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', style='Cluster', palette='deep', s=100)
plt.title('Customer Segmentation Visualization')    
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
```

```python
#5 clusters
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(X)
df['Cluster_5'] = kmeans.labels_  # Assign new cluster labels to the DataFrame
```

```python
# Visualize the new clusters
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster_5', style='Cluster_5', palette='deep', s=100)
plt.title('Customer Segmentation with 5 Clusters')  
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster_5')
plt.grid(True)
plt.show()

```

```python
# Visualize 5 clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster', palette='viridis', s=100)
plt.title('Customer Segments based on Income and Spending Score (5 Clusters)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```
